# How SuperResolution works?

SuperResolution is an image transformation technique with the help of which we can improve the quality of image and recover high resolution image from a given low resolution image. It allows us to remove the compression artifacts and transform the blurred images to sharper images by modifying the pixels. We can perform this task with the help of arcgis.learn.

## SuperResolution architecture

There is a wide variety of techniques available for Image SuperResolution but, we are using perceptual loss (feature loss) technique because it gives similar results as other techniques in the most efficient way. In this system, we are using two components: an image transformation network and a loss network.
<img src="data:image/PNG; base64, ">

We are using Unet as our image transformation network and VGG-16 as our network for feature loss.

To learn about Unet, you can refer to our guide [How Unet works?](https://developers.arcgis.com/python/guide/how-unet-works/).

1. Image transformation network (Unet): This network is parameterized by weights and takes the input images, transforms them by modifying pixels and generate the output image.

2. Loss Network (VGG-16): This network is pretrained on ImageNet data in which weights remain fixed during the training process. We use feature layers of this network to generate loss.

In the whole process, the low resolution image is fed into the image transformation network, which does the prediction $\hat{y}$. The predicted images $\hat{y}$ and the ground truth images y<sub>c</sub> are then fed into the loss network, where the loss between the two images is calculated using feature layers. Average of losses at the feature layers is taken and weights are updated during backpropagation.
